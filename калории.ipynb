```python
import re

def get_cap_from_string(string) -> int:
    result = 0
    target = string[4].strip()
    numbers = re.findall(r'\d+', target)
    if len(numbers) == 0:
        return 0
    
    cap = int(numbers[0])
    
    if target[-2:] == 'кг':
        result = cap * 10
    elif target[-1] == 'г' and target[-2] != 'к':
        result = cap / 100
    
    return result
    
```

```python
for x in data[:30]:
    print(get_cap_from_string(x))
```

```python
bludos = {}

for x in data:
    bludos[x[2]] = []
```

```python
for elem in bludos.keys():
    for x in data:
        if x[2] == elem:
            bludos[elem].append(x)
```

```python
result = {}
for bludo in bludos.keys():
    calories_final = 0
    for line in bludos[bludo]:
        try:
            calories_final += get_cap_from_string(line) * calories[line[3]]
        except:
            pass
    result[bludo] = calories_final
```
