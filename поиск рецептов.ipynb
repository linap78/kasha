```python
import csv
import random
list_recipe = []
ingr = input("Введите название продукта: ")
with open('ingredients_from_eda.ru.csv', encoding='utf-8') as r_file:
    data_reader = csv.DictReader(r_file)
    for line in data_reader:
        if ingr == line['ingredient']:
            list_recipe.append(line['name'])
random.shuffle(list_recipe)

print(list_recipe)
```
