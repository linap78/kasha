Какие блюда сложнее всего готовить, какие — легче всего

```python
import pandas as pd
```

```python
df = pd.read_csv("ingredients_from_eda.ru.csv") 
df
```

```python
recipes = df.groupby("name").agg({"ingredient": "count"}).sort_values(by="ingredient")
```

Наиболее легкие в приготовлении:

```python
recipes.head(10)
```

Наименее легкие в приготовлении:

```python
recipes.tail(10)
```
